## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

### Generate random dataset such that: 
<b>yi = a*xi + b (and a = -1, b = 2)</b>


```python
x_data = np.random.uniform(0, 20, 200)
```

```python
a = -1
b = 2

y_labels = a*x_data + b
```

```python
from sklearn.utils import shuffle 

x_data, y_labels = shuffle(x_data, y_labels)
```

#### Plot your data points. 

```python
plt.scatter(x_data, y_labels)
```

## Momentum-Based GD 

```python
def momentum_GD_LinearRegrission(data, labels, alpha = 0.001, gamma = 0.5, iteration = 50, batch_size = 4):
    theta_0  = 0
    theta_1  = 0
    v_theta0 = 0 
    v_theta1 = 0
    
    list_theta0 = []
    list_theta1 = []
    list_cost   = []
    list_yhat   = []
    
    len_data = len(labels)
    data_per_batch = len_data // batch_size

    for i in range(iteration):
        for j in range (batch_size):
            batch_data = data[j*data_per_batch : data_per_batch*(j+1)]
            batch_labels = labels[j*data_per_batch : data_per_batch*(j+1)]

            h = theta_0 + theta_1 * batch_data
            error = h - batch_labels
            cost = np.sum(error**2) / (2 * len(batch_data))

            gradient_theta0 = np.sum(error) / len(batch_data)
            gradient_theta1 = np.sum(error*batch_data) / len(batch_data)

            v_theta0 = gamma * v_theta0 + alpha * gradient_theta0
            v_theta1 = gamma * v_theta1 + alpha * gradient_theta1
            
            theta_0 = theta_0 - v_theta0
            theta_1 = theta_1 - v_theta1

        y_hat = theta_0 + theta_1 * data
        list_theta0.append(theta_0)
        list_theta1.append(theta_1)
        list_cost.append(cost)
        list_yhat.append(y_hat)
        #print("cost = ", cost)
    
    
    
    return list_theta0, list_theta1, list_cost, list_yhat, theta_0, theta_1, y_hat
    
```

```python
alpha  = 0.0001
epochs = 50
gamma  = 0.7

batch_size = 4
```

```python
list_theta0, list_theta1, list_cost, list_yhat, theta_0, theta_1, y_predicted = momentum_GD_LinearRegrission(x_data, y_labels, alpha = alpha, gamma = gamma, iteration = epochs, batch_size = batch_size)
```

```python
R_square = r2_score(y_labels, y_predicted)
print(R_square) 
```

```python
plt.plot(range(epochs), list_cost)
plt.title("Learning Process")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
plt.plot(list_theta0, list_cost)
plt.title("Change of theta 0 & loss")
plt.xlabel("Theta 0")
plt.ylabel("Loss")
```

```python
plt.plot(list_theta1, list_cost)
plt.title("Change of theta 1 & loss")
plt.xlabel("Theta 1")
plt.ylabel("Loss")
```

```python
plt.scatter(x_data, y_labels)
plt.plot(x_data, y_predicted, color='g')
plt.xlabel('X data')
plt.ylabel('Y labels')
plt.title('Best Fitted Line')
```

## NAG : Nesterov accelerated gradient

```python
def NAG_GD_LinearRegrission(data, labels, alpha = 0.001, gamma = 0.5, iteration = 50, batch_size = 4):
    theta_0  = 0
    theta_1  = 0
    v_theta0 = 0 
    v_theta1 = 0
    
    list_theta0 = []
    list_theta1 = []
    list_cost   = []
    list_yhat   = []
    
    len_data = len(labels)
    data_per_batch = len_data // batch_size

    for i in range(iteration):
        for j in range (batch_size):
            batch_data = data[j*data_per_batch : data_per_batch*(j+1)]
            batch_labels = labels[j*data_per_batch : data_per_batch*(j+1)]

            theta0_temp = theta_0 - gamma*v_theta0
            theta1_temp = theta_1 - gamma*v_theta1
            
            h = theta0_temp + theta1_temp * batch_data
            error = h - batch_labels
            cost = np.sum(error**2) / (2 * len(batch_data))
            
            gradient_theta0_temp = np.sum(error) / len(batch_data)
            gradient_theta1_temp = np.sum(error*batch_data) / len(batch_data)
            
            theta_0 = theta0_temp - alpha*gradient_theta0_temp
            theta_1 = theta1_temp - alpha*gradient_theta1_temp

            v_theta0 = gamma * v_theta0 + alpha * gradient_theta0_temp
            v_theta1 = gamma * v_theta1 + alpha * gradient_theta1_temp
            
        y_hat = theta_0 + theta_1 * data
        list_theta0.append(theta_0)
        list_theta1.append(theta_1)
        list_cost.append(cost)
        list_yhat.append(y_hat)
        #print("cost = ", cost)
    
    return list_theta0, list_theta1, list_cost, list_yhat, theta_0, theta_1, y_hat
    
```

```python
alpha  = 0.0001
epochs = 100
gamma  = 0.7

batch_size = 2
```

```python
list_theta0, list_theta1, list_cost, list_yhat, theta_0, theta_1, y_predicted = NAG_GD_LinearRegrission(x_data, y_labels, alpha = alpha, gamma = gamma, iteration = epochs, batch_size = batch_size)
```

```python
R_square = r2_score(y_labels, y_predicted)
print(R_square) 
```

```python
plt.plot(range(epochs), list_cost)
plt.title("Learning Process")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
plt.plot(list_theta0, list_cost)
plt.title("Change of theta 0 & loss")
plt.xlabel("Theta 0")
plt.ylabel("Loss")
```

```python
plt.plot(list_theta1, list_cost)
plt.title("Change of theta 1 & loss")
plt.xlabel("Theta 1")
plt.ylabel("Loss")
```

```python
plt.scatter(x_data, y_labels)
plt.plot(x_data, y_predicted, color='g')
plt.xlabel('X data')
plt.ylabel('Y labels')
plt.title('Best Fitted Line')
```
