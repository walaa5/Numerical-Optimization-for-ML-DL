## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generate random data such that: 
<b>yi = a*xi + b (and a = -1, b = 2)</b>


```python
x_data = np.random.uniform(0, 20, 200)
```

```python
a = -1
b = 2

y_labels = a*x_data + b
```

```python
from sklearn.utils import shuffle 

x_data, y_labels = shuffle(x_data, y_labels)
```

```python
plt.scatter(x_data, y_labels)
plt.title("data")
plt.xlabel("x_data")
plt.ylabel("labels")
```

## Adagrad

```python
def Adagrad_LinearRegrission(data, labels, alpha = 0.001, epsilon = 10**(-8), iteration = 50):
    
    #data = np.vstack(( np.ones(len(data)), data)).T
    data = np.column_stack((np.ones(len(data)), data))
    print(data.shape)
    thetas = np.zeros(data.shape[1])
    v = np.zeros(data.shape[1])
    
    list_thetas  = []
    list_cost    = []
    list_yhat    = []

    for i in range(iteration):

        h = data @ thetas
        error = h - labels
        cost = (error @ error) / (2 * len(data))
        gradient_thetas = (data.T @ error) / len(data)
        
        v = v + gradient_thetas ** 2 
        thetas = thetas - (alpha / (np.sqrt(v) + epsilon)) * gradient_thetas

        list_thetas.append(thetas)
        list_cost.append(cost)
        list_yhat.append(h)
    
    y_hat = data @ thetas
    
    return list_thetas, list_cost, list_yhat, thetas, y_hat
```

```python
alpha = 0.1
epochs = 400
epsilon = 10 ** (-8)

list_thetas, list_cost, list_yhat, thetas, y_predicted = Adagrad_LinearRegrission(x_data, y_labels, alpha = alpha, epsilon = epsilon, iteration = epochs)
```

```python
from sklearn.metrics import r2_score

R_square = r2_score(y_labels, y_predicted)
print(R_square)
```

```python
plt.plot(range(epochs), list_cost)
plt.title("Learning Process")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
plt.plot(np.asarray(list_thetas)[:, 0], list_cost)
plt.title("Change of theta 0 & loss")
plt.xlabel("Theta 0")
plt.ylabel("Loss")
```

```python
plt.plot(np.asarray(list_thetas)[:, 1], list_cost)
plt.title("Change of theta 1 & loss")
plt.xlabel("Theta 1")
plt.ylabel("Loss")
```

## RMSProp

```python
def RMSProp_LinearRegrission(data, labels, alpha = 0.001, epsilon = 10**(-8), beta=0.9, iteration = 50):
    
    #data = np.vstack((np.ones(len(data)), data)).T
    data = np.column_stack((np.ones(len(data)), data))
    thetas = np.zeros(data.shape[1])
    v = np.zeros(data.shape[1])
    
    list_thetas  = []
    list_cost    = []
    list_yhat    = []

    for i in range(iteration):

        h = data @ thetas
        error = h - labels
        cost = (error @ error) / (2 * len(data))
        gradient_thetas = (data.T @ error) / len(data)
        
        v = beta*v + (1-beta) * gradient_thetas ** 2 
        thetas = thetas - (alpha / (np.sqrt(v) + epsilon)) * gradient_thetas

        list_thetas.append(thetas)
        list_cost.append(cost)
        list_yhat.append(h)
    
    y_hat = data @ thetas
    
    return list_thetas, list_cost, list_yhat, thetas, y_hat
```

```python
alpha = 0.01
epochs = 200
epsilon = 10 ** (-8)
beta = 0.99

list_thetas, list_cost, list_yhat, thetas, y_predicted = RMSProp_LinearRegrission(x_data, y_labels, alpha = alpha, epsilon = epsilon, beta=beta, iteration = epochs)
```

```python
from sklearn.metrics import r2_score

R_square = r2_score(y_labels, y_predicted)
print(R_square)
```

```python
plt.plot(range(epochs), list_cost)
plt.title("Learning Process")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
plt.plot(np.asarray(list_thetas)[:, 0], list_cost)
plt.title("Change of theta 0 & loss")
plt.xlabel("Theta 0")
plt.ylabel("Loss")
```

```python
plt.plot(np.asarray(list_thetas)[:, 1], list_cost)
plt.title("Change of theta 1 & loss")
plt.xlabel("Theta 1")
plt.ylabel("Loss")
```

```python
# plot only the first 40 epoch 
n = 0
fig, axis = plt.subplots(5,8, figsize=(15,10))
fig.tight_layout()

for i in range(5):
    for j in range(8):
        axis[i, j].scatter(x_data, y_labels)
        axis[i, j].plot(x_data, list_yhat[n], color='g')
        axis[i, j].set_title("epoch {}".format(n+1))
        n = n+1
```

## Adam

```python
def Adam_LinearRegrission(data, labels, alpha = 0.001, epsilon = 10**(-8), beta1=0.9, beta2 = 0.999, iteration = 100, batch_size = 4):
    
    #data = np.vstack((np.ones(len(data)), data)).T
    data = np.column_stack((np.ones(len(data)), data))
    thetas = np.zeros(data.shape[1])
    v = np.zeros(data.shape[1])
    m = np.zeros(data.shape[1])
    t = 1                                           # for bias correction
    
    list_thetas  = []
    list_cost    = []
    list_yhat    = []
    
    len_data = len(labels)
    data_per_batch = len_data // batch_size

    for i in range(iteration):
        for j in range (batch_size):
            batch_data = data[j*data_per_batch : data_per_batch*(j+1)]
            batch_labels = labels[j*data_per_batch : data_per_batch*(j+1)]

            h = batch_data @ thetas
            error = h - batch_labels
            cost = (error @ error) / (2 * len(batch_data))
            gradient_thetas = (batch_data.T @ error) / len(batch_data)

            m = beta1 * m + (1-beta1) * gradient_thetas       # for momentum
            v = beta2*v + (1-beta2) * gradient_thetas ** 2    # for rms
            
            # bias correction part
            m_hat = m / (1 - beta1**t)
            v_hat = v / (1 - beta2**t)
            t = t+1

            thetas = thetas - (alpha / (np.sqrt(v_hat) + epsilon)) * m_hat
        
            
            
            list_thetas.append(thetas)
            list_cost.append(cost)
            list_yhat.append(h)

    y_hat = data @ thetas
    
    return list_thetas, list_cost, list_yhat, thetas, y_hat
```

```python
a = list(range(1))
print(a)
```

```python
alpha = 0.03
epochs = 100
batch_size = 1                     # len(y_labels)
beta1 = 0.9
beta2 = 0.9

list_thetas, list_cost, list_yhat, thetas, y_predicted = Adam_LinearRegrission(x_data, y_labels, alpha = alpha, epsilon = 10**(-8), beta1=beta1, beta2 = beta2, iteration = epochs, batch_size=batch_size)
```

```python
R_square = r2_score(y_labels, y_predicted)
print(R_square)
```

```python
plt.plot(range(epochs), list_cost)
plt.title("Learning Process")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
plt.plot(np.asarray(list_thetas)[:, 0], list_cost)
plt.title("Change of theta 0 & loss")
plt.xlabel("Theta 0")
plt.ylabel("Loss")
```

```python
plt.plot(np.asarray(list_thetas)[:, 1], list_cost)
plt.title("Change of theta 1 & loss")
plt.xlabel("Theta 1")
plt.ylabel("Loss")
```

```python
plt.scatter(x_data, y_labels)
plt.plot(x_data, y_predicted, color='g')
plt.xlabel('X data')
plt.ylabel('Y labels')
plt.title('Best Fitted Line')
```
